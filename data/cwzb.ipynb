```python
import os
os.chdir(r"D:\paper\rowdata")
import pandas as pd
import numpy as np
```

资产负债

```python
zcfz = pd.read_csv('ZCFZ.csv', parse_dates=[1], infer_datetime_format=True,
                   dtype={'Stkcd':str})
zcfz = zcfz.loc[:, ['Stkcd', 'Accper', 'F011201A']]

zcfz_ls = []

years = ['2009','2010','2011','2012','2013',
         '2014','2015','2016','2017','2018',
         '2019','2020','2021']

for year in years:
    date = year + '-12-31'
    zcfz_ls.append(zcfz[zcfz["Accper"]==date])

zcfz_1 = pd.concat(zcfz_ls)
zcfz_1
```

相对价值

```python
xdjz = pd.read_csv('XDJZ.csv', parse_dates=[1], infer_datetime_format=True,
                   dtype={'Stkcd':str})

xdjz_ls = []

years = ['2009','2010','2011','2012','2013',
         '2014','2015','2016','2017','2018',
         '2019','2020','2021']

for year in years:
    date = year + '-12-31'
    xdjz_ls.append(xdjz[xdjz["Accper"]==date])

xdjz_1 = pd.concat(xdjz_ls)
xdjz_1
```

重命名

```python
zcfz = zcfz_1.rename(columns = {"Stkcd": "Code", "Accper": "Time", "F011201A": "Zcfz"})
zcfz
```

```python
gdzc = gdzc_1.rename(columns = {"Stkcd": "Code", "Accper": "Time", "F030801A": "Gdzc"})
gdzc
```

```python
yszz = yszz_1.rename(columns = {"Stkcd": "Code", "Accper": "Time", "F081602C": "Yszz"})
yszz
```

```python
roa = roa_1.rename(columns = {"Stkcd": "Code", "Accper": "Time", "F050203B": "Roa"})
roa
```

```python
xdjz = xdjz_1.rename(columns = {"Stkcd": "Code", "Accper": "Time",
                                "F100802A": "Sz?", "F100901A": "TobinQ", "F101001A": "Zmszb"})
xdjz
```

合并数据

```python
data = pd.merge(zcfz, gdzc, on=['Code', 'Time'], how='outer')
data
```

```python
data = pd.merge(data, yszz, on=['Code', 'Time'], how='outer')
data
```

```python
data = pd.merge(data, roa, on=['Code', 'Time'], how='outer')
data
```

```python
data = pd.merge(data, xdjz.drop(['Sz?'], axis=1), on=['Code', 'Time'], how='outer')
data
```
