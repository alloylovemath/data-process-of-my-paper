```python
import pandas as pd
import numpy as np

import os
os.chdir(r'D:\paper\rowdata\rzrq_v')
```

```python
rzrq = pd.read_csv("rzrq_j.csv", dtype={"Mtyear":int})
rzrq
```

```python
Rzye = [np.sum(rzrq[rzrq["Mtyear"]==i]["Financebalance"]) for i in range(2010, 2022)]
Rqye = [np.sum(rzrq[rzrq["Mtyear"]==i]["Shortbalance"]) for i in range(2010, 2022)]
rzrq = [np.sum(rzrq[rzrq["Mtyear"]==i]["Mtbalance"]) for i in range(2010, 2022)]
```

```python
df = pd.DataFrame({"Year": [i for i in range(2010, 2022)], "Rzye": Rzye, "Rqye": Rqye, "Rzrqye": rzrq})
df
```
