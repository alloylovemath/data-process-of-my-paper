## 资产负债表相关数据

```python
import os
os.chdir(r"D:\paper\rowdata")
import pandas as pd
import numpy as np
```

```python
balance = pd.read_csv("balance.csv",parse_dates=[1], infer_datetime_format=True,
                     dtype={'Stkcd':str})
balance.head(20)
```

现有数据中包含季度数据、中报数据、年报数据等，我们只需要年报数据

```python
balance_ls = []

years = ['2009','2010','2011','2012','2013',
         '2014','2015','2016','2017','2018',
         '2019','2020','2021']

for year in years:
    date = year + '-12-31'
    balance_ls.append(balance[balance["Accper"]==date])
```

```python
balance_ls[-1]
```

balance_ls中是各年度相关数据，我们将其叠置起来

```python
balance_1 = pd.concat(balance_ls)
```

---

```python
balance_1 = balance_1.drop(['Typrep'], axis=1)
```

```python
balance_1
```

列名重命名：

```python
balance_1 = balance_1.rename(columns={"Stkcd": "Code",
                       "Accper": "Time", "A001121000": "Qtyskje",
                       "A001212000": "Gdzcje", "A001000000": "Zcze",
                       "A002120000": "Qtyfk", "A002000000": "Fzze"})
balance_1
```

```python
balance_1.to_csv('balance_clean.csv', index=False)
```
