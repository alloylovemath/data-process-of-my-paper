## 持股数据--holdings

```python
import os
os.chdir(r"D:\paper\rowdata")
import pandas as pd
import numpy as np
```

```python
holdings = pd.read_csv('holdings.csv', parse_dates=[1], infer_datetime_format=True,
                     dtype={'Stkcd':str})
holdings
```

```python
holdings["Reptdt"]
```

```python
holdings_ls = []

years = ['2009','2010','2011','2012','2013',
         '2014','2015','2016','2017','2018',
         '2019','2020','2021']

for year in years:
    date = year + '-12-31'
    holdings_ls.append(holdings[holdings["Reptdt"]==date])

holdings_1 = pd.concat(holdings_ls)
holdings_1
```

```python
holdings_2 = pd.read_csv('holdings_2.csv', parse_dates=[1], infer_datetime_format=True,
                     dtype={'Symbol':str})

holdings_2_ls = []

years = ['2009','2010','2011','2012','2013',
         '2014','2015','2016','2017','2018',
         '2019','2020','2021']

for year in years:
    date = year + '-12-31'
    holdings_2_ls.append(holdings_2[holdings_2["Enddate"]==date])

holdings_2_1 = pd.concat(holdings_2_ls)
holdings_2_1
```

---

重命名：

```python
holdings = holdings_1.rename(columns={"Stkcd": "Code", "Reptdt": "Time",
                                      "Shrcr1": "FirstDummy", "Shrs": "Two-ten"})
holdings
```

```python
holdings_2_1 = holdings_2_1.drop(['ShortName'], axis=1)

holdings_2 = holdings_2_1.rename(columns={"Symbol": "Code", "Enddate": "Time",
                                          "SharesBalance": "Gqzhd", "InsInvestorProp": "Jgtzz",
                                          "ConcurrentPosition": "Lzhy", "Mngmhldn": "Glccg",
                                          "Boardsize": "Dshgm", "IndDirectorRatio": "Dldszb"})
holdings_2
```

```python
holdings_2["Time"] = holdings_2["Time"].astype('datetime64')
```

```python
data = pd.merge(holdings, holdings_2, on=['Code', 'Time'], how='outer')
data
```

```python
data.dropna()
```

```python
data.to_csv('holdings_clean.csv', index=False)
```
