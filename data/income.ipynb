```python
import os
os.chdir(r"D:\paper\rowdata")
import pandas as pd
import numpy as np
```

```python
income = pd.read_csv('income.csv', parse_dates=[1], infer_datetime_format=True,
                    dtype={'Stkcd':str})
income.head(20)
```

只需要年报数据：

```python
income_ls = []

years = ['2009','2010','2011','2012','2013',
         '2014','2015','2016','2017','2018',
         '2019','2020','2021']

for year in years:
    date = year + '-12-31'
    income_ls.append(income[income["Accper"]==date])
    
income_1 = pd.concat(income_ls)
```

---

删除报表类型一列

```python
income_1 = income_1.drop(['Typrep'], axis=1)

income_1
```

---

列名重命名：

```python
income_1 = income_1.rename(
    columns = {"Stkcd": "Code", "Accper": "Time",
               "B001101000": "Yysr", "B001300000": "Yylr",
               "B002000000": "Jlr"}
)

income_1
```

```python
income_1.to_csv("income_clean.csv", index=False)
```
