```python
import os
os.chdir(r'D:\paper\rowdata')

import pandas as pd
import numpy as np

import sqlite3 as sq3
```

创建database

```python
paper_Con = sq3.connect('paper_d.db')
cursor = paper_Con.cursor()
```

# basics

```python
os.chdir(r'D:\paper\rowdata\basics')
```

```python
basics = pd.read_csv('basics_clean.csv', dtype = {"Code": str})
basics
```

将所有数据转化为str:

```python
for i in range(8):
    basics.iloc[:, i] = basics.iloc[:, i].astype(str)
```

创建表格，插入数据：

```python
os.chdir(r'D:\paper\rowdata')

query = '''
CREATE TABLE basics(
Code TEXT, Industry_1 TEXT, Industry_2 TEXT,
Estbdt TEXT, Listdt TEXT, Province TEXT,
City TEXT, Jyxz TEXT, PRIMARY KEY(Code)
);
'''

cursor.execute(query)
paper_Con.commit()
```

```python
n = len(basics["Code"])
data = [tuple(basics.iloc[i, :]) for i in range(n)]
data
```

```python
stmt = '''
INSERT INTO basics VALUES(?, ?, ?, ?, ?, ?, ?, ?)
'''

cursor.executemany(stmt, data)

paper_Con.commit()
```

basics解决了
