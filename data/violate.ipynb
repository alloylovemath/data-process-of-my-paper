```python
import os
os.chdir(r"D:\paper\rowdata\violate")
import pandas as pd
import numpy as np
```

```python
violate = pd.read_csv('violate.csv',
                     dtype={'Symbol':str})
violate = violate.iloc[:, [1, 3, 4]]
violate
```

```python
vi_code = ['P2508', 'P2509', 'P2510', 'P2514']
vi_code_wide = ['P2508', 'P2509', 'P2510', 'P2514', 'P2599']
```

```python
def vi_dec(s1, ls):
    s = s1.split(',')
    for item in s:
        if item in ls:
            return True
    return False
```

```python
violate["Vio"] = [1 if vi_dec(i, vi_code) else 0 for i in violate["ViolationTypeID"]]
violate["Vio_wide"] = [1 if vi_dec(i, vi_code_wide) else 0 for i in violate["ViolationTypeID"]]
violate
```

```python
violate = violate.drop('ViolationTypeID', axis=1)
violate = violate.rename(columns={"Symbol": "Code", "DeclareDate": "Time_a"})
violate
```

```python
code_ls = []
with open("Code.txt", 'r') as code:
    for line in code:
        code_ls.append(line[:-1])

date_ls = ['2009-12-31', '2010-12-31', '2011-12-31',
          '2012-12-31', '2013-12-31', '2014-12-31',
          '2015-12-31', '2016-12-31', '2017-12-31',
          '2018-12-31', '2019-12-31', '2020-12-31',
          '2021-12-31',]

n_1 = len(code_ls)
n_2 = len(date_ls)

Code_ls = []
for code in code_ls:
    for i in range(n_2):
        Code_ls.append(code)

Time_ls = []
for i in range(n_1):
    for date in date_ls:
        Time_ls.append(date)

df = pd.DataFrame({"Code": Code_ls, "Time_a": Time_ls})
df
```

```python
violate["Year"] = [i[:4] for i in violate["Time_a"]]
violate
```

```python
def vio_not(code, year):
    try:
        df1 = violate[violate["Code"]==code]
        n = df1[df1["Year"]==year]["Vio"].sum()
        if n > 0:
            return True
        else:
            return False
    except:
        return False

def vio_not_wide(code, year):
    try:
        df1 = violate[violate["Code"]==code]
        n = df1[df1["Year"]==year]["Vio_wide"].sum()
        if n > 0:
            return True
        else:
            return False
    except:
        return False
```

```python
n = len(df["Code"])
df["Vio"] = ['Y' if vio_not(df["Code"].iloc[i], df["Time_a"].iloc[i][:4]) else 'N' for i in range(n)]
df["Vio_wide"] = ['Y' if vio_not_wide(df["Code"].iloc[i], df["Time_a"].iloc[i][:4]) else 'N' for i in range(n)]
df
```

```python
def vio_not_num(code, year):
    try:
        df1 = violate[violate["Code"]==code]
        n = df1[df1["Year"]==year]["Vio"].sum()
        if n > 0:
            return n
        else:
            return 0
    except:
        return 0

def vio_not_wide_num(code, year):
    try:
        df1 = violate[violate["Code"]==code]
        n = df1[df1["Year"]==year]["Vio_wide"].sum()
        if n > 0:
            return n
        else:
            return 0
    except:
        return 0
```

```python
n = len(df["Code"])
df["NumVio"] = [vio_not_num(df["Code"].iloc[i], df["Time_a"].iloc[i][:4]) for i in range(n)]
df["NumVio_wide"] = [vio_not_wide_num(df["Code"].iloc[i], df["Time_a"].iloc[i][:4]) for i in range(n)]
df
```
