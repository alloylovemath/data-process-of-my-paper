```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
smoke = pd.read_excel('Smoking.xlsx',index_col=None)
smoke.head(5)
```

---

(a)

```python
smoke_a_all = smoke['smoker']
smoke_a_no_ban = smoke[smoke['smkban']==0]['smoker']
smoke_a_ban = smoke[smoke['smkban']==1]['smoker']
```

```python
result_a = pd.DataFrame(
    {'p':[smoke_a_all.mean(), smoke_a_no_ban.mean(), smoke_a_ban.mean()],
     'SEM':[smoke_a_all.sem(), smoke_a_no_ban.sem(), smoke_a_ban.sem()]},
    index=['All Worker', 'No Smoking Ban', 'Smoking Ban'])
result_a.round(3)
```

---

(b) linear probability model

```python
import statsmodels.api as sm
```

```python
model_linear = sm.OLS(smoke['smoker'],
                      sm.add_constant(smoke['smkban'])).fit()
print(model_linear.summary())
```

---

(c)

```python
X = smoke.iloc[:, 1:]
X['age**2'] = X['age']**2
X.head()
```

```python
model_linear_2 = sm.OLS(smoke['smoker'], sm.add_constant(X)).fit()
```

```python
model_linear_2.summary()
```

(f) probit

```python
from statsmodels.discrete.discrete_model import Probit
```

```python
model_probit = Probit(smoke['smoker'], sm.add_constant(X)).fit()
```

```python
model_probit.summary()
```

logit

```python
from statsmodels.discrete.discrete_model import Logit
```

```python
model_logit = Logit(smoke['smoker'], sm.add_constant(X)).fit()
model_logit.summary()
```
